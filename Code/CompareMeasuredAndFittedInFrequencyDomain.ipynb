# Convert fitted field to FFT

```python
f_sample = 500
```

```python
fitted_data = {}
```

```python
nu = 5
```

```python
fitdatafilename = '../Results/Exp2/Pickles/FittedData_{}Hz.pk'.format(nu)
```

```python
with open(fitdatafilename,'rb') as file_obj:
    fitted_data[nu] = pickle.load(file_obj)
```

```python
FFTdatafilename = '../Res'
```

```python
fitted_data[nu]['time'] = (720-fitted_data[nu]['theta'])/(360*nu)
```

```python
Excess_Noise = {}
Excess_Noise['AV']={}
Excess_Noise['AV']['X'] = {}
Excess_Noise['AV']['X'][nu] = fitted_data[nu]['measured field']['AV']['X'] - fitted_data[nu]['calculated field']['AV']['X']

Excess_Noise['AW']={}
Excess_Noise['AW']['Y'] = {}
Excess_Noise['AW']['Y'][nu] = fitted_data[nu]['measured field']['AW']['Y'] - fitted_data[nu]['calculated field']['AW']['Y']
```

```python
plt.plot(fitted_data[nu]['theta'],fitted_data[nu]['measured field']['AV']['X'],label = 'measured')
plt.plot(fitted_data[nu]['theta'],fitted_data[nu]['calculated field']['AV']['X'],label = 'fit')
plt.plot(fitted_data[nu]['theta'],Excess_Noise['AV']['X'][nu],label = 'residual')
plt.legend()
plt.grid()
plt.title('AV X')
plt.xlabel('rotation angle (°)')
plt.ylabel('Magnetic field (pT)')
plt.savefig(SavePlotDir_Exp2+'compare_time_domain.png',dpi=300,bbox_inches = 'tight')

# plt.plot(fitted_data[nu]['theta'],Excess_Noise['AW']['Y'][nu])
```

```python
fitted_data_FFT={}
fitted_data_FFT[nu] = {}
fitted_data_FFT[nu]['AW'] = {}
fitted_data_FFT[nu]['AV'] = {}
```

```python
# TimeArrayLength = len(fitted_data[nu]['time'])
# fitted_data_FFT[nu]['AW']['Z'] = (np.fft.rfft(fitted_data[nu]['calculated field']['AW']['Z'])/TimeArrayLength)
# fitted_data_FFT[nu]['AW']['Y'] = (np.fft.rfft(fitted_data[nu]['calculated field']['AW']['Y'])/TimeArrayLength)
# fitted_data_FFT[nu]['AV']['X'] = (np.fft.rfft(fitted_data[nu]['calculated field']['AV']['X'])/TimeArrayLength)
# fitted_data_FFT[nu]['AV']['Y'] = (np.fft.rfft(fitted_data[nu]['calculated field']['AV']['Y'])/TimeArrayLength)
# fitted_data_FFT[nu]['frequency']  = f_sample/TimeArrayLength*np.arange(1,int(TimeArrayLength/2)+2,1)
```

```python
N_rep = 1
```

```python
t = np.tile(fitted_data[nu]['time'],N_rep)
Calc_AVX_timedomain = np.tile(fitted_data[nu]['calculated field']['AV']['X'],N_rep)
Meas_AVX_timedomain = np.tile(fitted_data[nu]['measured field']['AV']['X'],N_rep)
Excess_AVX_timedomain = np.tile(Excess_Noise['AV']['X'][nu],N_rep)
# Excess_AWY_timedomain = np.tile(Excess_Noise['AW']['Y'][nu],200)

size_t = len(t)
FFT_Calc_AVX = np.fft.rfft(Calc_AVX_timedomain)/size_t
FFT_Meas_AVX = np.fft.rfft(Meas_AVX_timedomain)/size_t
FFT_Excess_AVX = np.fft.rfft(Excess_AVX_timedomain)/size_t
# FFT_Excess_AWY = np.fft.rfft(Excess_AWY_timedomain)/size_t
freq = f_sample/size_t*np.arange(1,int(size_t/2)+2,1)
```

```python
plt.loglog(freq,abs(FFT_Meas_AVX),alpha = 0.8,label = 'measured')
plt.loglog(freq,abs(FFT_Calc_AVX),alpha = 0.8,label = 'fitted')
plt.loglog(freq,abs(FFT_Excess_AVX),alpha = 0.8,label = 'residual')
# plt.loglog(freq,abs(FFT_Excess_AWY),alpha = 0.8)
# plt.ylim(1e-2,1e1)
plt.grid()
plt.grid(which='minor',linestyle='--')
plt.legend(loc ='best')
plt.title('AV X')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnetic field (pT)')
plt.savefig(SavePlotDir_Exp2+'compare_FFT.png',dpi=300,bbox_inches = 'tight')

# plt.xlim(0.8*nu,13*nu)
```

```python
plt.loglog(freq,abs(FFT_Meas_AVX),alpha = 0.8,label = 'measured')
plt.loglog(freq,abs(FFT_Calc_AVX),alpha = 0.8,label = 'fitted')
plt.loglog(freq,abs(FFT_Excess_AVX),alpha = 0.8,label = 'residual')
# plt.loglog(freq,abs(FFT_Excess_AWY),alpha = 0.8)
plt.ylim(1e-30,1e-14)
plt.grid()
plt.grid(which='minor',linestyle='--')
plt.legend(loc ='best')
plt.title('AV X')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnetic field (pT)')
# plt.xlim(0.8*nu,13*nu)
plt.savefig(SavePlotDir_Exp2+'compare_FFT_zoom_bottom.png',dpi=300,bbox_inches = 'tight')
```
